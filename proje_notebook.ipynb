## Senaryo 8: Isparta Afet Toplanma AlanlarÄ± Drone Rota Optimizasyonu

Bu projede, Isparta AFAD Ä°l MÃ¼dÃ¼rlÃ¼ÄŸÃ¼ tarafÄ±ndan belirlenen 10 resmi afet toplanma alanÄ± arasÄ±nda drone ile paket taÅŸÄ±macÄ±lÄ±ÄŸÄ± iÃ§in en kÄ±sa rotayÄ± KarÄ±nca Kolonisi AlgoritmasÄ± (ACO) kullanarak buluyoruz.

**Koordinatlar:** Isparta AFAD Ä°l MÃ¼dÃ¼rlÃ¼ÄŸÃ¼'nden alÄ±nmÄ±ÅŸtÄ±r.


## 1. Gerekli KÃ¼tÃ¼phanelerin Ä°Ã§e AktarÄ±lmasÄ±


````python
import numpy as np
import matplotlib.pyplot as plt
from data.coordinates import ISPARTA_AFET_ALANLARI, get_location_names, get_coordinates
from core.matrix_utils import haversine_distance
from core.ant_algorithm import ACO

print("âœ… KÃ¼tÃ¼phaneler baÅŸarÄ±yla yÃ¼klendi!")
````

## 2. Isparta Afet Toplanma AlanlarÄ± KoordinatlarÄ±

AFAD Ä°l MÃ¼dÃ¼rlÃ¼ÄŸÃ¼ tarafÄ±ndan belirlenen resmi toplanma noktalarÄ±:


````python
location_names = get_location_names()
coordinates = get_coordinates()

print("ğŸ¢ Isparta AFAD Afet Toplanma AlanlarÄ±:\n")
print("-" * 60)
for i, (name, coord) in enumerate(zip(location_names, coordinates), 1):
    print(f"{i:2d}. {name:40s} | {coord}")
print("-" * 60)
print(f"\nğŸ“Š Toplam Nokta SayÄ±sÄ±: {len(location_names)}")
````

## 3. Mesafe Matrisinin OluÅŸturulmasÄ±

Haversine formÃ¼lÃ¼ ile noktalar arasÄ± kuÅŸ uÃ§uÅŸu mesafeler hesaplanÄ±r:


````python
n = len(coordinates)
distance_matrix = np.zeros((n, n))

print("ğŸ”„ Mesafe matrisi hesaplanÄ±yor...\n")

for i in range(n):
    for j in range(n):
        if i != j:
            distance_matrix[i][j] = haversine_distance(coordinates[i], coordinates[j])

print(f"âœ… Mesafe matrisi boyutu: {distance_matrix.shape}")
print(f"\nğŸ“ Ã–rnek Mesafeler:")
print(f"   {location_names[0][:30]} â†’ {location_names[1][:30]}: {distance_matrix[0][1]:.2f} km")
print(f"   {location_names[0][:30]} â†’ {location_names[5][:30]}: {distance_matrix[0][5]:.2f} km")
````

## 4. KarÄ±nca Kolonisi AlgoritmasÄ±nÄ±n Ã‡alÄ±ÅŸtÄ±rÄ±lmasÄ±

ACO parametreleri ayarlanarak optimizasyon baÅŸlatÄ±lÄ±r:


````python
# Algoritma Parametreleri
n_ants = 20
n_iterations = 100
alpha = 1.0
beta = 2.0
evaporation_rate = 0.5
Q = 100

print("ğŸœ KarÄ±nca Kolonisi AlgoritmasÄ± BaÅŸlatÄ±lÄ±yor...")
print(f"   KarÄ±nca SayÄ±sÄ±: {n_ants}")
print(f"   Ä°terasyon SayÄ±sÄ±: {n_iterations}")
print(f"   Alpha (Î±): {alpha}")
print(f"   Beta (Î²): {beta}")
print(f"   BuharlaÅŸma OranÄ±: {evaporation_rate}\n")

# ACO AlgoritmasÄ±nÄ± Ã‡alÄ±ÅŸtÄ±r
aco = ACO(distance_matrix, n_ants, n_iterations, alpha, beta, evaporation_rate, Q)
best_path, best_distance, history = aco.run()

print("\nğŸ‰ Optimizasyon TamamlandÄ±!")
print("=" * 70)
print(f"ğŸ¯ En Ä°yi Mesafe: {best_distance:.2f} km")
print("=" * 70)
print(f"\nğŸ—ºï¸ Optimal Drone RotasÄ±:\n")
for i, idx in enumerate(best_path):
    if i < len(best_path) - 1:
        print(f"   {i+1:2d}. {location_names[idx]}")
    else:
        print(f"   {i+1:2d}. {location_names[idx]} (DÃ¶nÃ¼ÅŸ)")
print("=" * 70)
````

## 5. YakÄ±nsama GrafiÄŸi

AlgoritmanÄ±n iterasyonlar boyunca nasÄ±l iyileÅŸtiÄŸini gÃ¶rselleÅŸtiriyoruz:


````python
plt.figure(figsize=(14, 6))

# Ana grafik
plt.plot(history, linewidth=2.5, color='#ff6b9d', label='En Ä°yi Mesafe')
plt.fill_between(range(len(history)), history, alpha=0.3, color='#ff6b9d')

# Grafik Ã¶zelleÅŸtirme
plt.xlabel('Ä°terasyon SayÄ±sÄ±', fontsize=14, fontweight='bold')
plt.ylabel('Mesafe (km)', fontsize=14, fontweight='bold')
plt.title('ğŸœ ACO AlgoritmasÄ± - YakÄ±nsama GrafiÄŸi', fontsize=16, fontweight='bold', pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=12)

# Ä°statistikler
plt.text(0.02, 0.98, f'BaÅŸlangÄ±Ã§: {history[0]:.2f} km\nSon: {history[-1]:.2f} km\nÄ°yileÅŸme: {history[0]-history[-1]:.2f} km', 
         transform=plt.gca().transAxes, fontsize=11, 
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print(f"\nğŸ“Š Ä°statistikler:")
print(f"   Ä°lk iterasyon: {history[0]:.2f} km")
print(f"   Son iterasyon: {history[-1]:.2f} km")
print(f"   Ä°yileÅŸtirme: {history[0] - history[-1]:.2f} km ({((history[0]-history[-1])/history[0]*100):.1f}%)")
````
